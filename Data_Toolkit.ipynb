2. How does broadcasting work in NumPy?
-> Broadcasting in NumPy allows us to perform arithmetic operations on arrays of different shapes without reshaping them. It automatically adjusts the smaller array to match the larger array's shape by replicating its values along the necessary dimensions. This makes element-wise operations more efficient by reducing memory usage and eliminating the need for loops. In this article, we will see how broadcasting works.

17. Explain the difference between apply() and map() in Pandas.
->
map():
Exclusively for Series.
Primarily used for element-wise transformations or substitutions of values, often with a dictionary, another Series, or a function that takes a single value and returns a single value.
Generally more optimized and faster for simple, one-to-one mappings on a Series.

apply():
Can be used on both Series and DataFrames.
On a Series, it applies a function to each element.
On a DataFrame, it applies a function along an axis (row-wise or column-wise), allowing for more complex operations involving multiple elements or columns.
Offers greater flexibility as the applied function can return a single value or a Series/DataFrame, enabling aggregations or transformations that change the data's shape.
Can be slower than map() for simple element-wise operations due to its overhead.

```python
#2 Write a Pandas script to find the mean of a specific column in a DataFrame.

import pandas as pd

# Create a sample DataFrame
data = {'Column_A': [10, 20, 30, 40, 50],
        'Column_B': [5, 15, 25, 35, 45],
        'Column_C': [1, 2, 3, 4, 5]}
df = pd.DataFrame(data)

# Specify the column for which to calculate the mean
column_name = 'Column_A'

# Calculate the mean of the specified column
mean_value = df[column_name].mean()

# Print the result
print(f"The mean of '{column_name}' is: {mean_value}")
```

```python
#4 How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# --- 1. Create Sample Data ---
# Let's create a DataFrame with some numerical data
# We'll generate 5 columns and 100 rows of random data.
# Some columns will have a positive/negative correlation to demonstrate.

np.random.seed(42) # for reproducibility

# Independent variables
data = {
    'Feature_A': np.random.rand(100) * 10,
    'Feature_B': np.random.randn(100) * 5 + 10,
    'Feature_C': np.random.randint(0, 100, 100),
}

# Create some correlated features
# Feature_D will be positively correlated with Feature_A
data['Feature_D'] = data['Feature_A'] * 0.8 + np.random.randn(100) * 2

# Feature_E will be negatively correlated with Feature_B
data['Feature_E'] = 20 - (data['Feature_B'] * 0.7 + np.random.randn(100) * 3)

df = pd.DataFrame(data)

print("--- Sample DataFrame Head ---")
print(df.head())
print("\n--- Sample DataFrame Info ---")
df.info()
print("\n" + "="*50 + "\n")

# --- 2. Calculate the Correlation Matrix ---
# The .corr() method on a Pandas DataFrame calculates the pairwise correlation
# of columns. By default, it uses Pearson correlation.
correlation_matrix = df.corr()

print("--- Correlation Matrix ---")
print(correlation_matrix)
print("\n" + "="*50 + "\n")

# --- 3. Visualize the Correlation Matrix with Seaborn Heatmap ---
plt.figure(figsize=(10, 8)) # Set the size of the plot

sns.heatmap(
    correlation_matrix,
    annot=True,      # Show the correlation values on the heatmap
    cmap='coolwarm', # Choose a colormap. 'coolwarm' is good for showing positive/negative correlations.
                     # Other options: 'viridis', 'plasma', 'magma', 'Blues', 'Greens', etc.
    fmt=".2f",       # Format the annotations to two decimal places
    linewidths=.5,   # Add lines between cells for better separation
    cbar=True,       # Show the color bar
    square=True      # Ensure the cells are square for better visual representation
)

plt.title('Correlation Matrix Heatmap', fontsize=16) # Add a title to the plot
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
plt.yticks(rotation=0)             # Keep y-axis labels horizontal
plt.tight_layout()                 # Adjust layout to prevent labels from being cut off
plt.show()

```

```python
#8 Create a line plot with multiple lines using Matplotlib.
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]     # First line
y2 = [1, 2, 3, 4, 5]       # Second line
y3 = [25, 16, 9, 4, 1]     # Third line

# Create the plot
plt.plot(x, y1, label='y = x²', color='blue', linestyle='-')
plt.plot(x, y2, label='y = x', color='green', linestyle='--')
plt.plot(x, y3, label='y = reverse(x²)', color='red', linestyle=':')

# Add titles and labels
plt.title("Multiple Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
